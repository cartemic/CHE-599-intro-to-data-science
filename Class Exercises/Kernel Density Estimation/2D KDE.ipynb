```julia
using CSV
using DataFrames
using PyPlot
using ScikitLearn
using Statistics
using Random
using LaTeXStrings
using Printf
using Seaborn

# (optional) check out all of the styles! https://matplotlib.org/3.1.1/gallery/style_sheets/style_sheets_reference.html
PyPlot.matplotlib.style.use("bmh")

# (optional)change settings for all plots at once, e.g. font size
rcParams = PyPlot.PyDict(PyPlot.matplotlib."rcParams")
rcParams["font.size"] = 16;
```

```julia
df = CSV.read(
    "old_faithful.dat", 
    datarow=27, 
    header=26, 
    delim=" ", 
    ignorerepeated=true
)

first(df, 6)
```

```julia
plt, ax = subplots(1, 1, figsize=(18, 8))
ax.scatter(df[!, :eruption], df[!, :waiting], marker="x")
ax.set_xlabel("Eruption Time (min)", weight="bold")
ax.set_ylabel("Waiting Time (min)", weight="bold")
```

```julia
plt, ax = subplots(1, 1, figsize=(18, 8))
kdeplot(df[!, :eruption], df[!, :waiting], bw="scott", ax=ax, shade=true)
ax.scatter(df[!, :eruption], df[!, :waiting], marker="x")
ax.set_xlabel("Eruption Time (min)", weight="bold")
ax.set_ylabel("Waiting Time (min)", weight="bold")
```
